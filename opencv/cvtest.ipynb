```python
img = cv2.imread("nj.png")
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
```

```python
img3 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(img3)
plt.show()
```

```python
img4 = cv2.bilateralFilter(img3, 9, 75, 75)
plt.imshow(img4)
plt.show()
```

```python
img5 = cv2.adaptiveThreshold(img4, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 4)
plt.imshow(img5)
plt.show()
```

```python
img6 = cv2.medianBlur(img5, 13)
plt.imshow(img6)
plt.show()
```
